# Lampiran 5: Potongan kode preprocessing (Python/Notebook)

```python
# Import library
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load data
df = pd.read_csv('/content/cm1.csv')
df = df.dropna()

# Encode target
df['defects'] = df['defects'].astype(int)

# Pisahkan fitur & target
X = df.drop(columns='defects')
y = df['defects']

# Normalisasi
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Gabungkan
df_final = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)
```

# Revisi: lakukan proses imbalance data, analisis korelasi, bandingkan minimal 3 algoritma, cari yang lebih akurat

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# Upload dan load data
from google.colab import files

df = pd.read_csv('cm1.csv')
df = df.dropna()
df['defects'] = df['defects'].astype(int)

# Normalisasi
X = df.drop(columns='defects')
y = df['defects']

scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
df_final = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_final.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Fitur')
plt.show()

# Korelasi dengan target
corr_target = df_final.corr()['defects'].abs().sort_values(ascending=False)
print("\nKorelasi Fitur terhadap Target 'defects':\n", corr_target)
```

# Imbalance Handling pakai SMOTE

```python
!pip install imblearn

from imblearn.over_sampling import SMOTE

X = df_final.drop(columns='defects')
y = df_final['defects']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Terapkan SMOTE
sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

print("\nDistribusi label setelah SMOTE:")
print(y_train_sm.value_counts())
```

# Membandingkan 3 Algoritma

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000)
}

results = {}

for name, model in models.items():
    model.fit(X_train_sm, y_train_sm)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    print(f"\n{name} - Akurasi: {acc:.2f}")
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f'Confusion Matrix - {name}')
    plt.show()

    results[name] = acc

# Bandingkan akurasi
print("\nPerbandingan Akurasi Model:")
for k, v in results.items():
    print(f"{k}: {v:.2f}")

# Model terbaik
best_model = max(results, key=results.get)
print(f"\nModel dengan akurasi terbaik adalah: {best_model}")
```
